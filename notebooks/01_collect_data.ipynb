```python
from google_play_scraper import reviews, Sort
import pandas as pd
```

```python
result, _ = reviews(
    'com.gojek.app',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=5000
)

print(f"Berhasil ambil {len(result)} ulasan")
```

```python
df = pd.DataFrame(result)[['content', 'score']]
print(f"Shape: {df.shape}")
df.head()
```

```python
df = df.dropna(subset=['content'])
print(f"Setelah hapus NaN: {df.shape}")
```

```python
def convert_label(score):
    if score <= 2:
        return 0  # negatif
    elif score >= 4:
        return 1  # positif
    else:
        return None  # rating 3 dibuang

df['label'] = df['score'].apply(convert_label)
df = df.dropna(subset=['label'])
df['label'] = df['label'].astype(int)
print(f"Setelah labeling: {df.shape}")
print(df['label'].value_counts())
```

```python
df[['content', 'label']].to_csv('../data/raw/reviews.csv', index=False)
print("✅ Data tersimpan di data/raw/reviews.csv")
```
